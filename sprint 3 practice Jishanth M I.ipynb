```python
import pandas as pd
import numpy as np

data=pd.read_csv(r"C:\Users\Jishanth M I\Downloads\DS1_C4_S3_SalesData_Practice.csv")
data
```

```python
# Task 1
#Sort the cata in ascending order of sales and display the result


# Sort the DataFrame by 'sales' column in ascending order
sorted_data = data.sort_values(by='Sales')

print(sorted_data)
```

```python
# Task 2
#Display all the rows where the sale amount is 45,000

sale_amount = 45000

filtered_data = data[data['Sales'] >= sale_amount]

filtered_data
```

```python
# Task 3
#Filter the data separately for the states of California. Texas, and Montana and save them in three different data frames Sort the three data frames in increasing order of sale amount and display them separately.


# Filter data for each state
california_data = data[data['State'] == 'California'].sort_values(by='Sales')
texas_data = data[data['State'] == 'Texas'].sort_values(by='Sales')
montana_data = data[data['State'] == 'Montana'].sort_values(by='Sales')

print("California Data:")
print(california_data)

print("\nTexas Data:")
print(texas_data)

print("\nMontana Data:")
print(montana_data)
```

```python
# Task 4
#Calculate the sum of sales (of all six months) for the states specified in Task 3 and display the state-wise total sales
#Sample of output: Total sales in the state of California: 11,000
        
        
California=california_data['Sales'].sum()
print("Total sale in California: ",California)

Texas=texas_data['Sales'].sum()
print("Total sale in Texas: ",Texas)

Montana=montana_data['Sales'].sum()
print("Total sale in Montana: ",Montana)
```

```python
# Task 5
#Display the months with the highest and the lowest monthly sale amount, for the stores in the state of California


# Filter data for stores in California
ca_stores = data[data['State'] == 'California']

monthly_sales = ca_stores.groupby(ca_stores['Month'])['Sales'].sum()

# Find the month with the highest and lowest sales
highest_month = monthly_sales.idxmax()
lowest_month = monthly_sales.idxmin()


print("Month with the highest sales:", highest_month)
print("Month with the lowest sales:", lowest_month)
```

```python
# Task 6
#Create and display a new data frame consisting of state- wise total sales as shown in the output below.
#Hint: Refer to the Jupyter notebook of the Concept session to figure out how ow to create a data frame ming lists, Populate two lists with name of the state and total sale amount.


state=data.groupby('State')['Sales'].sum()

sale_details=pd.DataFrame({"State":state.index,"Sales":state.values})
sale_details
```

```python
# Task 7
#Sort the rows in the data frame created in Task 6 by the sale amount. Save the data frame as a total sales.csv file in your local system. Check your current working directory to see if the file is saved correctly


total_sales=sale_details.sort_values('Sales')
total_sales

total_sales.to_csv("total_sales.csv")
```
