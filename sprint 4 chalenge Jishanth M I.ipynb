```python
import pandas as pd
import numpy as np

data=pd.read_csv(r"C:\Users\Jishanth M I\Downloads\DS1_C4_S4_Used_Car_Prices_Data_Challenge.csv")
data
```

```python
# Task 1
# Load the data into a data frame and check for missing values. Drop the rows with missing values and proceed with the cleaned- up data for the subsequent tasks.



# Display the count of missing values for each column
missing_values = data.isnull().sum()
print("Missing Values:\n", missing_values)

# Drop rows with any missing values
cleaned_data = data.dropna()

# Display the shape of the cleaned data
print("\nShape of Cleaned Data:", cleaned_data.shape)
```

```python
# Task 2
# Thomas, a potential customer from California, is interested in buying a low-priced car, manufactured in 2014. John needs to identify 100 lowest priced cars from the year 2014 with CA as the state of registration. Help John create and display the details of potential cars.


# Filter for cars with the specified year and state
filtered_cars = data[(data['year'] == 2014) & (data['state'] == 'ca')]

# Sort the filtered cars by price in ascending order
sorted_cars = filtered_cars.sort_values(by='mp',ascending=True)[:100]

sorted_cars
```

```python
# Task 3
# Thomas is confused looking at the long list of 100 cars. Help John to display the output in a cross table based on the car manufacturers and prices.


# Use the crosstab function to create the cross table
cross_table = pd.crosstab(index=data['make'], columns='mp', values=data['mp'], aggfunc='mean')

print(cross_table)
```

```python
# Task 4
# Thomas is particularly interested in Chevrolet cars. Eric is willing to offer a 10% discount on Chevrolet cars from the list of 100 cars identified in the previous tasks.
#Display the updated data with only Chevrolet cars with their original prices and discounted prices.
    
    
# Filter Chevrolet cars and calculate discounted prices
chevrolet_cars = sorted_cars[sorted_cars['make'] == 'Chevrolet']
chevrolet_cars['discounted_price'] = chevrolet_cars['mp'] * 0.9

# Display the updated data
print(chevrolet_cars[['model', 'mp', 'discounted_price']])
```

```python
# Task 5
# Thomas and several other customers seem to be interested in the cars with black or gray colored interiors. John wants to figure out if the cars with interiors in black or gray are priced higher or lower than cars having interiors of other colors.
#Determine the average price of all the 100 lowest priced cars from the year 2014 based on their interior colors.


#Filter cars from the year 2014
year_2014_cars = data[data['year'] == 2014]

# Sort by price in ascending order
lowest_priced_cars = year_2014_cars.sort_values(by='mp', ascending=True)[:100]

# Calculate the average price for each interior color category
average_price_by_color = lowest_priced_cars.groupby('interior')['mp'].mean()

print(average_price_by_color)
```

```python
# Task 6
# Most buyers seem to prefer white and black colored cars. John wants to analyze color-wise availability of cars, for each manufacturer from the years 2010-15.


# Filter data for the years 2010-2015
filtered_data = data[(data["year"] >= 2010) & (data["year"] <= 2015)]

# Group data by manufacturer and color, and then count occurrences
color_counts = filtered_data.groupby(['make', 'color']).size().reset_index(name='Count')

color_counts
```

```python
# Task 7
# Eric, another customer is interested in a good condition BMW Sedan manufactured between 2010 and 2015. John wants to display the details of all such cars that have a condition score >=4.5. Help John to figure out of the year-wise maximum expected price (mp) of the above BMW cars manufactured between the year 2010 and 2015 grouped by each model.


# Filter cars based on condition and year
filtered_cars = data[(data['condition'] >= 4.5) & (data['year'].between(2010, 2015))]

# Group by model and year, and find the max expected price
max_expected_prices = filtered_cars.groupby(['model', 'year'])['mp'].max()

max_expected_prices
```

```python
# Task 8
#John wants to recommend Eric to also consider SUVS from other manufacturers before finalizing the car. 
#John needs to list the top 100 SUVS available between the years 2010 and 2015, based on their prices (in descending order) with an odometer reading <= 30000.


# Filter SUVs between 2010 and 2015 with an odometer reading <= 30000
filtered_suv_data = data[(data['year'] >= 2010) & (data['year'] <= 2015) & (data['odometer'] <= 30000)]

# Sort the filtered SUVs by price in descending order
sorted_suv_data = filtered_suv_data.sort_values(by='mp', ascending=False).head(100)

sorted_suv_data
```
