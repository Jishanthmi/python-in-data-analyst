```python
import pandas as pd
import numpy as np

data=pd.read_csv(r"C:\Users\Jishanth M I\Downloads\DS1_C4_S4_Test_Scores_Data_Practice.csv")
data
```

```python
# Task 1
# Load the data from the file into a data frame and identify the missing values (if any) using the command dataframe.isnull().sum().
#Treat the missing values with the most frequently found data in that column. The dataframe['column_name'].value_counts()
#command will help you identify the most frequently found data for any column with missing values.

Hint: Replace test1 missing values with the mean of test1 scores and test2 missing values with the median of test2 scores.
missing_values = data.isnull().sum()
missing_values

data['gender'].fillna(data['gender'].mode()[0],inplace=True)
data['test1'].fillna(data['test1'].mean(),inplace=True)
data['test2'].fillna(data['test2'].mean(),inplace=True)

print(missing_values)
```

```python
# Task 2
# Identify and display the duplicate rows (if any). Delete the duplicate rows in the data.

data[data.duplicated(keep=False)]
# no duplicated element
```

```python
# Task 3
#In the column named 'tuitionfee_waiver', replace the value 'Does not qualify' and 'Qualifies for reduced/free' with No and Yes, respectively. Display the updated column data and verify if the values are replaced.

data['tuitionfee_waiver']= data['tuitionfee_waiver'].replace({'Does not qualify':'No'})
data['tuitionfee_waiver']= data['tuitionfee_waiver'].str.replace('Qualifies for reduced/free','Yes')
print(data['tuitionfee_waiver'])
```

```python
# Task 4
#Calculate and display the average of test1 scores of students at each school. Display the name of the school that has the highest average test1 score.
#Hint: Use the "group by" operation to first segregate the data for each school and then calculate the average test score.
 
    
# Group by 'School' and calculate the average test1 score
average_scores = data.groupby('school')['test1'].mean()
highest_avg_score = average_scores.max()

print("Average Test1 Scores: ",average_scores)
print("\nSchool with the Highest Average Test1 Score: ",highest_avg_score)
```

```python
# Task 5
#Display the highest test2 score of each school in a tabular format.
#Display the name of the top five schools based on the output of the above task.
#Hint: Use the pivot table function.


pivot_table = pd.pivot_table(data,index='school', values='test2', aggfunc='max').sort_values('test2', ascending=False)[:5]

print(pivot_table)
```

```python
# Task 6
#The school named 'FBUMG' is offering a scholarship of $500 for the students of class '5LQ' who have scored>=80 in test2.
#Display the information of these students along with student_ID and scholarship amounts in a tabular format.
#Hint: Use the concat operation to append a new column to your data frame. The scholarship column will have to be populated as a series, prior to using it in the concat operation.
    

filtered_data = data[(data['classroom'] == '5LQ') & (data['test2'] >= 80)]

# Calculate scholarship amounts
filtered_data['scholarship'] = 500

# Display the information in a tabular format
print(filtered_data[['student_id', 'scholarship']])
```

```python
# Task 7
#Display the count of each type of teaching method, segregated by school setting.


# Create a cross table
cross_table = pd.crosstab(index=data['school_setting'], columns=data['teaching_method'])

print(cross_table)
```

```python
# Task 8
#Display the gender-wise count of students in different school settings - rural, suburban, and urban.
#Hint: Use the "group by" operation on school setting and gender and then use the "count" operation.
    
    
# Grouping and counting
gender_count = data.groupby(['school_setting', 'gender']).count()

print(gender_count)
```

```python
# Task 9
#Display the average test2 scores of male and female students from each school setting.


# Group by school_setting and gender, then calculate the mean test2_score
result = data.groupby(['school_setting', 'gender'])['test2'].mean()

print(result)
```
