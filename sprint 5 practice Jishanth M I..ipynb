```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_excel(r"C:\Users\Jishanth M I\Downloads\DS1_C4_S5_Employee_Data_Practice.xlsx")
df
```

```python
# Task 1
#What is the total annual salary for all the employees of each department? Display the result using a bar graph.
#Hint: 1) Summarize the department-wise sum of annual salaries.
#ii) Use index and values properties of the summarized data to plot the bar graphs.



department_salaries = df.groupby('Department')['Annual Salary ($)'].count()
plt.bar(department_salaries.index, department_salaries.values,color='green')
plt.xlabel('Department')
plt.ylabel('Total Annual Salary')
plt.title('Total Annual Salary by Department')
plt.show()
```

```python
# Task 2
#The company wants to analyze gender equality across several departments. Display the department-wise total number of male and female employees using a DataFrame.
#Hints: 1. Summarize the department- and gender-wise count of employees.
#2. Create DataFrame with all department names and the respective male and female counts.



# Group the data by Department and Gender, and then count the occurrences
department_gender_counts = df.groupby(['Department', 'Gender']).size().unstack(fill_value=0)

# Create a bar plot
department_gender_counts.plot(kind='bar', stacked=True)
plt.xlabel('Department')
plt.ylabel('Number of Employees')
plt.title('Gender Equality Across Departments')

plt.show()
```

```python
# Task 3
#What salary do you think the company might offer a new employee (a candidate with no work experience) of the IT department?
#Hint: i) Apply a filter on the data where work experience is 0 and the department is IT.
#ii) Calculate the average of salaries of filtered data.


# Filter data where work experience is 0 and the department is IT
filtered_data = df[(df["Work_Experience"] == 0) & (df["Department"] == "IT")]

# Calculate the average salary
average_salary = filtered_data["Annual Salary ($)"].mean()
average_salary

# Create a bar chart
plt.bar(["Annual Salary ($)"], [average_salary])
plt.ylabel("Salary")
plt.title("Average Salary for New Employees with No Work Experience in IT Department")
plt.show()
```

```python
# Task 4
#Display the department-wise ratio for the sum of annual salaries for all employees in each department, using a pie chart.
#Hint: i) Summarize data (annual salary) by department.
#ii) Use index and values properties of the summarized data to plot the pie chart.
#iii) Set the 'autopct' attribute value of the pie chart to %.2f %%.


# Group data by department and calculate the sum of annual salaries
department_salaries = df.groupby('Department')['Annual Salary ($)'].count()

# Plotting the pie chart
plt.figure(figsize=(8, 6))
plt.pie(department_salaries, labels=department_salaries.index, autopct='%.2f %%')
plt.title('Department-wise Ratio of Sum of Annual Salaries')
plt.show()
```

```python
# Task 5
#Display the distribution of employee count by age group. using a histogram.
#Hint: Plot age group buckets (bins) on the x-axis and number of employees on the y axis. Choose the bin size as 25-30, 30-35, and so on, all the way up to 50.
    
    
# Define age group bins
age_bins = np.arange(25, 55, 5)

# Create the histogram
plt.hist(x=df['Age'], bins=age_bins,rwidth=0.98,color='green')

# Add labels and title
plt.xlabel('Age Group')
plt.ylabel('Number of Employees')
plt.title('Distribution of Employee Count by Age Group')
plt.show()
```

```python
# Task 6
#Display the salary trend based on employee experience using a scatterplot.


# Create a scatterplot
plt.figure(figsize=(10, 6))
plt.scatter(x='Work_Experience',y='Annual Salary ($)', color='green', marker='o', label='Salary Trend')

# Adding labels and title
plt.title('Salary Trend Based on Employee Experience')
plt.xlabel('Experience (Years)')
plt.ylabel('Salary ($)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Task 7
#Determine the department-wise spread of all employees' salaries using a separate box plot for each department. 
#Identify and write down the names of the departments in which there are outliers in employee salary.


plt.figure(figsize=(10, 6))
df.boxplot(column='Annual Salary ($)', by='Department')
plt.title('Department-wise Spread of Employee Salaries')
plt.xlabel('Department')
plt.ylabel('Salary')

plt.show()
```
