```python
import pandas as pd
import numpy as np
```

```python
# Task 1
#Display the details of the customers whose spending score is greater than 80. Also, display the total count of these customers.

ecommerce_data = np.genfromtxt(r"C:\Users\Jishanth M I\Downloads\DS1_C4_S1_Shopping_Data_Challenge.csv", delimiter=',',skip_header=1)

#Finding the customers with spending score greater than 88 
spending_score= ecommerce_data[ecommerce_data[:, 3]> 80]

# Getting customer details
customer_details= spending_score[:, :3]

#Getting count of customers 
count= customer_details.shape[0]

print("Details of customers with spending score greater than 80:", customer_details) 
print("Count of customers with spending score greater than 80:", count)
```

```python
# Task 2
#Extract and display the data of the customers who are 20 to 24 years of age. Note: This data helps analyze age-related expenditures.

#Finding customers between 20 and 24 years old 
over_20_to_24= ecommerce_data[(ecommerce_data[:, 1] >= 20) & (ecommerce_data[:, 1] <= 24)]

# Getting customer details 
Customer_ID= over_20_to_24[:, 0]

# Getting number of employees 
num_employees= Customer_ID.shape[0]

print("Number of customers between 20 and 24 years old:", num_employees) 
print("Customers between 20 and 24 years old:", Customer_ID)
```

```python
# Task 3
#Display the IDs of the customers with the lowest annual income(s) and the highest annual income.


#Finding customers with Lowest and highest annual incomes
lowest_annual_income = ecommerce_data[ecommerce_data[:, 2] == np.min(ecommerce_data[:, 2])] 
highest_annual_income= ecommerce_data[ecommerce_data[:, 2] == np.max(ecommerce_data[:, 2])]

# Getting customer IDs
lowest_annual_income_IDs = lowest_annual_income[:, 0] 
highest_annual_income_IDs= highest_annual_income[:, 0]

print("IDs of customers with the lowest annual income:", lowest_annual_income_IDs) 
print("IDs of customers with the highest annual income:", highest_annual_income_IDs)
```

```python
# Task 4
#Display the details of the customers who are less than 45 years of age, have an annual income greater than $35K, and have a spending score greater than 75. How many such customers are there in the given data?


customer_details = ecommerce_data[(ecommerce_data[:, 1] < 45) & (ecommerce_data[:, 2]> 35) & (ecommerce_data[:, 3]> 75)]

#Displaying the CustomerIDS
Customer_ID=customer_details[:, 0]

#Count of customer_IDS
count=customer_details.shape[0]

print("IDs of customers based on criteria:", Customer_ID)
print("Number of customers:", count)
```
