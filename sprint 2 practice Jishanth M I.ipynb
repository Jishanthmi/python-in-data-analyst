```python
import pandas as pd
import numpy as np
```

```python
# Task 1
#Create a Pandas Series variable to assign unique student IDs to the 40 students. The student IDs must start from the number 1001, increasing by 1 for each student. Display the student IDs of all the students  

student_ids = pd.Series(np.arange(1001,1041,1))

print(student_ids)
```

```python
# Task 2
#The physics and mathematics scores of 40 students is provided in the form of a list in the Jupyter notebookStore the scores in two separate Pandas Series variables and display them.
#Note: The student IDs generated in Task 1 will become the row index for this task.
    
maths_marks=[35, 86, 63, 90, 50, 63, 15, 72, 15, 54, 72, 83, 58, 70, 92, 40, 59, 63, 53, 59, 55, 85, 18, 83, 67, 85, 46, 59, 77, 71, 77, 56, 79, 61, 82, 76, 60, 52, 88, 78]
physics_marks=[85, 94, 63, 59, 76, 64, 87, 55, 51, 71, 54, 94, 33, 72, 75, 67, 85, 15, 50, 50, 19, 55, 59, 65, 76, 13, 57, 70, 51, 94, 55, 50, 71, 91, 75, 51, 90, 70, 87, 56]

maths_series = pd.Series(maths_marks, index=student_ids)
physics_series = pd.Series(physics_marks, index=student_ids)
print("Maths marks", maths_series)
print("physics marks", physics_series)
```

```python
# Task 3
#Create and display Pandas Series variable containing the sum of scores of physics and mathematics, along with the student ID's

total_scores = maths_series + physics_series

print(total_scores)
```

```python
# Task 4
#The cut-off score to qualify for round 1 is 120 marks out of a maximum marks of 200. This includes the marks of both mathematics and physics.
#List down the student IDs of the qualifying students.

qualifying_students = total_scores[total_scores > 120]

print(qualifying_students)
```

```python
# Task 5
#Create a data frame containing the sum of math and physics scores of all 40 students, where the sum of score exceeds 120. The column name in this data frame must be "round1 qualifiers".
#Note: The row index/label must contain the student IDs of the 40 students generated in Task 1.


qualifiers_df = pd.DataFrame({"round1 qualifiers": qualifying_students})

print(qualifiers_df)
```

```python
# Task 6
#Create a data frame of the students who qualify for round 2. The conditions to qualify for round 2 are as follows Minimum score of 50 in math.
#Minimum score of 50 in physics.
#Sum of maths and physics scores is 120.
#The column name in the data frame should be "round 2 qualifiers". The data values in this column should be the student IDs of the qualifying students.


round2_qualifiers = physics_series[(physics_series >= 50) & (maths_series >= 50)]

round2_df = pd.DataFrame({"round 2 qualifiers": round2_qualifiers})

print(round2_df)
```
