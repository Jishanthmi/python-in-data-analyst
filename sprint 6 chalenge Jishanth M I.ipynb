# Task 1

```python
# Most people prefer houses that are at least 6.000 square feet and are comer plot houses. Write a query to fetch these rows and push them into a new table in MySQL

filter_data=data[(data['LotArea']>=6000 )&(data['LotConfig']=='Corner')].loc[:,['LotArea','LotConfig']]
filter_data
```

```python
filter_data.to_sql('TASK1',eng,index=True)
```

# Task 2

```python
# Most people seem to prefer houses that are one family and one story, preferably on flat land avoiding mountains or valleys/banks. Write a query to fetch these rows and push them into a new table in MySQL.


filtered_data=data[(data['HouseStyle']=='1Story')& (data['BldgType']=='1Fam')&(data['LandContour']!='Bnk')&(data['LandContour']!='HLS')]
filtered_data
```

```python
filtered_data.to_sql("TASK2",eng,index=True)
```

# Task 3

```python
# Compact townhouses are hot-selling properties in California Write a query to fetch these rows and push them into a new table in MySQL


filter_data=data[(data['BldgType']=='TwnhsE')|(data['BldgType']=='Twnhsl')]
dat=filter_data.groupby('HouseStyle').count()
dat
```

```python
dat.to_sql("TASK3",eng,index=False)
```

# Task 4

```python
# Houses with 3+ bedrooms, a basement, and a cement exterior are popular among most customers Write a query to fetch these rows and push them into a new table in MySQL.


filter_data=data[(data['Exterior1st']=='CemntBd')&(data['BsmtCond']!='NA')&(data['BsmtFinSF1']!='NA')&(data['BedroomAbvGr']>=3)]
filter_data
```
