```python
# Task 1
#Identify the top 10 car manufacturers based on the city mileage and display the result using a horizontal bar graph with the manufacturer on the y-axis and mileage on the x-axis.
#Do you notice any outliers in the above output? If yes, what is the reason and how would you resolve it?


# Group by manufacturer and calculate average city mileage
manufacturer_mileage = df.groupby('Make')['City_Mileage_km_litre'].mean()

# Sort manufacturers by average city mileage and select top 10
top_manufacturers = manufacturer_mileage.sort_values(ascending=False).head(10)

# Plotting
plt.figure(figsize=(10, 6))
plt.bar(top_manufacturers.index, top_manufacturers.values, color='green')
plt.xlabel('City Mileage')
plt.ylabel('Manufacturer')
plt.title('Top 10 Car Manufacturers by City Mileage')
plt.show()
```

```python
# Task 4
#Display the car mileages for the following four companiesusing a pivot table: Hyundai, Mahindra, Renault, and Skoda
#Create 4 separate pivot tables capturing the following details: 1. Mean of city mileage 2. Median of city mileage 3. Mean of highway mileage 4. Median of highway mileage
    
    

# Extracting the relevant columns for the selected car manufacturers
selected_companies = ['Hyundai', 'Mahindra', 'Renault', 'Skoda']
selected_columns = ['Make', 'City_Mileage_km_litre', 'Highway_Mileage_km_litre']

# Filtering the data for the selected car manufacturers
df_selected = df[df['Make'].isin(selected_companies)][selected_columns]

# Create pivot tables for each of the requested details
pivot_mean_city = df_selected.pivot_table(index='Make', values='City_Mileage_km_litre', aggfunc='mean')
pivot_median_city = df_selected.pivot_table(index='Make', values='City_Mileage_km_litre', aggfunc='median')
pivot_mean_highway = df_selected.pivot_table(index='Make', values='Highway_Mileage_km_litre', aggfunc='mean')
pivot_median_highway = df_selected.pivot_table(index='Make', values='Highway_Mileage_km_litre', aggfunc='median')

plt.figure(figsize=(10, 6))

# Mean City Mileage
pivot_mean_city.plot(kind='bar', color='skyblue')
plt.title('Mean City Mileage')
plt.ylabel('Mean Mileage')

# Median City Mileage
pivot_median_city.plot(kind='bar', color='green')
plt.title('Median City Mileage')
plt.ylabel('Median Mileage')

# Mean Highway Mileage
pivot_mean_highway.plot(kind='bar', color='orange')
plt.title('Mean Highway Mileage')
plt.ylabel('Mean Mileage')

# Median Highway Mileage
pivot_median_highway.plot(kind='bar', color='purple')
plt.title('Median Highway Mileage')
plt.ylabel('Median Mileage')

plt.show()
```

```python
# Task 5
# Display the mean and median values of city mileages of the following manufacturers, using a stacked column chart. Hyundai, Renault, Mahindra, and Skoda


selected_manufacturers = ['Hyundai', 'Mahindra', 'Renault', 'Skoda']
df_selected = df[df['Make'].isin(selected_manufacturers)]

# Group by manufacturer and calculate mean and median city mileages
grouped_data = df_selected.groupby('Make')['City_Mileage_km_litre'].agg(['mean', 'median'])

plt.figure(figsize=(10, 6))
grouped_data.plot(kind='bar',stacked=True, color=['skyblue', 'orange'])
plt.xlabel('Manufacturer')
plt.ylabel('City Mileage')
plt.title('Mean and Median City Mileage Comparison')

plt.show()
```

```python
# Task 6
# Display the average mileage (city and highway) of all manufactures using a horizonal bar chart, in ascending order of mileage.
#Note: The missing values in the mileage column need to be filled with the mean value.


# Calculate mean mileage values
mean_city_mileage = df["City_Mileage_km_litre"].mean()
mean_highway_mileage = df["Highway_Mileage_km_litre"].mean()

# Fill missing values with mean values
df["City_Mileage_km_litre"].fillna(mean_city_mileage, inplace=True)
df["Highway_Mileage_km_litre"].fillna(mean_highway_mileage, inplace=True)

# Calculate average mileage for each manufacturer
avg_mileage = df.groupby("Make")[["City_Mileage_km_litre", "Highway_Mileage_km_litre"]].mean()
avg_mileage["Average_Mileage"] = avg_mileage.mean(axis=1)
avg_mileage = avg_mileage.sort_values(by="Average_Mileage")

# Create a horizontal bar chart
plt.figure(figsize=(10, 10))
avg_mileage["Average_Mileage"].plot(kind="barh", color="skyblue")
plt.xlabel("Average Mileage")
plt.ylabel("Manufacturer")
plt.title("Average Mileage of Manufacturers")
plt.show()
```
